```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el archivo Excel
df = pd.read_excel('Student_performance_data-_.xlsx')

# Limpiar los nombres de las columnas para eliminar espacios
df.columns = df.columns.str.strip()

# Crear un gráfico de líneas de 'Age' vs 'GPA'
plt.figure(figsize=(10, 6))
df.plot(x='Age', y='GPA', kind='line', marker='o', linestyle='-')
plt.title('Relación entre Edad y Promedio de Calificaciones (GPA)')
plt.xlabel('Edad')
plt.ylabel('Promedio de Calificaciones (GPA)')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el archivo Excel
df = pd.read_excel('Student_performance_data-_.xlsx')

# Histograma de la Edad
plt.figure(figsize=(10, 6))
df['Age'].hist(bins=10, color='skyblue', edgecolor='black')
plt.title('Distribución de Edades de los Estudiantes')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.grid(False)
plt.show()
```

```python
# Calcular GPA promedio por género
gpa_gender = df.groupby('Gender')['GPA'].mean()

# Gráfico de barras
plt.figure(figsize=(10, 6))
gpa_gender.plot(kind='bar', color=['blue', 'pink'])
plt.title('Promedio de GPA por Género')
plt.xlabel('Género')
plt.ylabel('GPA Promedio')
plt.xticks(rotation=0)
plt.grid(False)
plt.show()
```

```python
# Gráfico de dispersión
plt.figure(figsize=(10, 6))
plt.scatter(df['StudyTimeWeekly'], df['GPA'], alpha=0.5, color='orange')
plt.title('Relación entre Horas de Estudio Semanal y GPA')
plt.xlabel('Horas de Estudio Semanal')
plt.ylabel('GPA')
plt.grid(True)
plt.show()
```

```python
# Gráfico de caja
plt.figure(figsize=(12, 6))
df.boxplot(column='GPA', by='ParentalEducation', grid=False)
plt.title('Distribución de GPA por Nivel de Educación de los Padres')
plt.suptitle('')
plt.xlabel('Nivel de Educación de los Padres')
plt.ylabel('GPA')
plt.xticks(rotation=45)
plt.grid(False)
plt.show()
```

```python
# Gráfico de pastel
plt.figure(figsize=(8, 8))
df['Ethnicity'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Distribución Étnica de los Estudiantes')
plt.ylabel('')  # Para quitar el label 'Ethnicity'
plt.grid(False)
plt.show()
```
