```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Crear el DataFrame
data = {
    'A\u00f1o': [2021, 2022, 2023, 2024],
    'Velocidad Viento Min (m/s)': [0, 0, 0, 0.1],
    'Velocidad Viento M\u00e1x (m/s)': [4.9, 4.8, 5.0, 4.8],
    'Promedio Velocidad Viento (m/s)': [1.2, 1.1, 1.3, 1.4],
    'Desviaci\u00f3n Est\u00e1ndar Velocidad Viento': [0.8, 0.8, 0.8, 0.9],
    'Direcci\u00f3n Viento Min (\u00b0)': [0, 0, 0, 0],
    'Direcci\u00f3n Viento M\u00e1x (\u00b0)': [359, 360, 360, 360],
    'Temperatura Promedio (\u00b0C)': [15.3, 15.1, 15.7, 16.1]
}

df = pd.DataFrame(data)

# Configurar el estilo de seaborn
sns.set_style("whitegrid")

# Crear la figura y los ejes
fig, ax = plt.subplots(figsize=(12, 8))

# Graficar las l\u00edneas
sns.lineplot(data=df, x='A\u00f1o', y='Velocidad Viento Min (m/s)', marker='o', label='Vel. Viento Min')
sns.lineplot(data=df, x='A\u00f1o', y='Velocidad Viento M\u00e1x (m/s)', marker='o', label='Vel. Viento M\u00e1x')
sns.lineplot(data=df, x='A\u00f1o', y='Promedio Velocidad Viento (m/s)', marker='o', label='Promedio Vel. Viento')
sns.lineplot(data=df, x='A\u00f1o', y='Temperatura Promedio (\u00b0C)', marker='o', label='Temperatura Promedio')

# Configurar el t\u00edtulo y las etiquetas
plt.title('Datos de Viento y Temperatura (2021-2024)', fontsize=16)
plt.xlabel('A\u00f1o', fontsize=12)
plt.ylabel('Valores', fontsize=12)

# Ajustar la leyenda
plt.legend(fontsize=10, loc='center left', bbox_to_anchor=(1, 0.5))

# Ajustar los m\u00e1rgenes
plt.tight_layout()

# Guardar la figura como PNG
plt.savefig('datos_viento_temperatura.png', dpi=300, bbox_inches='tight')

print("El gr\u00e1fico ha sido guardado como 'datos_viento_temperatura.png'")

# Mostrar el gr\u00e1fico
plt.show()
```
